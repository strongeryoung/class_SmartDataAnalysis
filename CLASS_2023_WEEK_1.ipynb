## 6. 데이터 분석 도구

- 프로그램 언어
  - Python
    - 오픈소스 프로그래밍 언어
    - 다양하게 활용할 수 있는 방대한 라이브러리 제공    
      (EX : Numpy, Scipy, Pandas, Scikit-learn, matplotlib, seaborn 등)  

  - R
    - 오픈소스 통계용 프로그래밍 언어
    - 통계 및 비전공자에게 상대적으로 진입 장벽 낮은 언어
    - 무료 사용 가능
    - AWS 등에서 미지원 (사용에 한계)

  - SQL
    - RDBMS(Relational DataBase Management System)
    - DB에 저장된 데이터를 조회하고 관리하기 위해 만들어진 언어
    - python, R보다 직관적이고 저장된 데이터를 다양한 형태로 가공할 수 있어 데이터를 사용하는 다양한 분야에서 활용되고 있음

  - EXCEL
    - 가장 기본적인 데이터 분석 도구이며 1차 데이터 자료
    - 기초적인 자료관리 기능, 회계, 통계처리 기능 존재
    - 자동화 (매크로 기능) 탑재
    - 간단한 시각화 가능
