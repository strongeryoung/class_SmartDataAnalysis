# **1. Open API**

- Chat GPT가 설명하는 Open API란?

Open API(Open Application Programming Interface)는 소프트웨어 응용 프로그램들이 서로 상호 작용할 수 있도록 하는 프로그래밍 인터페이스를 의미합니다. 이는 서로 다른 소프트웨어 시스템이나 서비스가 특정한 방식으로 통신할 수 있게 해줍니다.

Open API의 주요 특징은 다음과 같습니다:

1. **개방성(Openness):** Open API는 공개적으로 사용 가능하며, 문서화되어 있어 다른 개발자들이 쉽게 이를 사용할 수 있습니다. 이는 제한된 액세스나 특정한 조건들 없이 누구나 사용할 수 있음을 의미합니다.

2. **표준화(Standardization):** Open API는 일반적으로 특정한 표준 또는 프로토콜을 따르며, 이를 통해 다른 시스템 간의 상호 운용성(interoperability)을 보장합니다.

3. **다양성(Diversity):** Open API는 다양한 응용 프로그램 및 플랫폼에서 사용될 수 있습니다. 예를 들어, 웹 기반 애플리케이션, 모바일 앱, 데스크톱 소프트웨어 등에서 활용할 수 있습니다.

4. **자유로운 접근(Free Access):** Open API를 사용하기 위해서는 일반적으로 무료 또는 저렴한 가격으로 이용할 수 있습니다. 이로써 많은 개발자들이 이를 활용하여 다양한 서비스나 애플리케이션을 개발할 수 있습니다.

5. **클라이언트-서버 통신(Client-Server Communication):** Open API는 클라이언트와 서버 간의 통신을 위한 명세를 제공합니다. 이를 통해 클라이언트는 서버로 데이터를 요청하고, 서버는 그에 대한 응답을 제공합니다.

예를 들어, Twitter의 Open API를 사용하면 다른 애플리케이션에서도 트위터와 상호 작용할 수 있습니다. 이를 통해 새로운 트윗을 게시하거나 타임라인을 가져오는 등의 작업이 가능합니다. Open API는 다양한 서비스와 응용 프로그램 간에 정보 및 기능을 공유하는 데 사용됩니다.

## 1.1 What is json ?


JSON은 JavaScript Object Notation의 약어로, 데이터를 효과적으로 표현하고 전달하기 위한 경량의 데이터 형식입니다. JSON은 사람과 기계 모두가 이해하기 쉬운 텍스트 기반의 데이터 형식입니다.

JSON은 일반적으로 키-값 쌍의 집합으로 이루어져 있습니다. 이러한 키-값 쌍은 객체(object)라고 불립니다. 객체는 중괄호 {}로 감싸져 있습니다. 각 키는 문자열이어야 하며, 값은 숫자, 문자열, 불리언, 배열, 다른 객체 등 다양한 데이터 타입일 수 있습니다.


```python
{
    "이름": "홍길동",
    "나이": 30,
    "성별": "남성",
    "주소": "서울 동대문구",
    "취미": "음악"
}
```

이 JSON 객체는 사람의 정보를 나타내고 있습니다. 예를 들어, "이름"은 키이고 "홍길동"은 값입니다. "주소"는 또 다른 객체이며, 그 안에 또 다른 키-값 쌍이 있습니다.

JSON은 웹에서 데이터를 주고받거나, 프로그램 간에 데이터를 교환할 때 주로 사용됩니다. 이는 서버와 클라이언트 간의 통신에서 매우 일반적이며, 다양한 프로그래밍 언어에서 JSON 데이터를 다루는 라이브러리와 함수를 제공합니다.

```python
import json

json_data = '''
{
    "이름": "홍길동",
    "나이": 30,
    "성별": "남성",
    "주소":  "서울 동대문구",
    "취미": "음악"
}
'''

# JSON 문자열을 파이썬 딕셔너리로 변환
data = json.loads(json_data)

# 데이터 확인
print(data)
```

## 1.2 What is XML ?

XML은 Extensible Markup Language의 약어로, 데이터를 표현하고 전송하기 위한 마크업 언어입니다. XML은 사용자가 자신만의 태그를 정의하여 데이터를 구조화할 수 있는 확장 가능한 특징을 가지고 있습니다.

간단히 말해, XML은 데이터를 계층적 구조로 표현하는 데 사용되며, 특정 목적이나 어플리케이션에 따라 자유롭게 정의된 태그들을 사용할 수 있습니다. 이는 다양한 종류의 데이터를 표현할 수 있도록 해줍니다.

<XML 예시>

```python
<bookstore>
  <book category="아마데우스">
    <title>책 제목 1</title>
    <author>저자 1</author>
    <price>10.99</price>
  </book>
  <book category="모짜르트">
    <title>책 제목 2</title>
    <author>저자 2</author>
    <price>12.99</price>
  </book>
  <book category="볼프강">
    <title>책 제목 3</title>
    <author>저자 3</author>
    <price>9.99</price>
  </book>
</bookstore>
```

```python
key = KEYS['ODP']

# 국가코드  중국 (112)
# 각 행의 컬럼, 이름, 값을 가지는 리스트 만들기
row_list = [] # 행값
name_list = [] # 열이름값
value_list = [] #데이터값

country_code = 275

nat_cd = str(country_code)
ED_cd = 'E'
dt_list =  ['202301', '202302', '202303', '202304', '202305', '202306', '202307', '202308']

for dt in dt_list:
    #print(dt)
    url = f'http://openapi.tour.go.kr/openapi/service/EdrcntTourismStatsService/getEdrcntTourismStatsList?YM={dt}&NAT_CD={nat_cd}&ED_CD={ED_cd}&serviceKey={key}'
    response = requests.get(url)
    data = response.text
    # data = json.loads(data)

    # 깔끔한 출력 위한 코드
    pp = pprint.PrettyPrinter(indent=4)

    ## 각 컬럼 값 ## (포털 문서에서 꼭 확인하세요)
    #bs4 사용하여 item 태그 분리
    xml_obj = bs4.BeautifulSoup(data,'lxml-xml')
    rows = xml_obj.findAll('item')
    # print(rows)

    # xml 안의 데이터 수집
    for i in range(0, len(rows)):
        # print(i)
        columns = rows[i].find_all()
        #첫째 행 데이터 수집
        for j in range(0,len(columns)):
            if i == 0:
                # 컬럼 이름 값 저장
                name_list.append(columns[j].name)
            # 컬럼의 각 데이터 값 저장
            value_list.append(columns[j].text)
        # 각 행의 value값 전체 저장
        row_list.append(value_list)

        # 데이터 리스트 값 초기화
        value_list=[]

data_frame = pd.DataFrame(row_list, columns=['항목', '구분', '코드', '국적', '입국자수', '나열순', '날짜'])
# 데이터 형태 확인
# data_frame.info()
data_frame['날짜']  = pd.to_datetime(data_frame['날짜'], format='%Y%m')+ pd.offsets.MonthEnd(1)
data_frame['입국자수'] = data_frame['입국자수'].astype('int64')
data_frame = data_frame.set_index('날짜')
```

이 코드는 특정 기간의 연도와 월을 조합하여 날짜를 생성하는 함수를 정의하는 파이썬 코드입니다. 함수는 `start` 연도부터 `end` 연도 전까지의 모든 월을 포함하는 날짜 목록을 생성합니다. 그리고 `end_month`까지의 월까지만 포함하는 마지막 연도를 처리합니다.

이제 각 부분을 자세히 설명하겠습니다:

1. `def module_for_datelist(start, end, end_month):`:
   - 함수 `module_for_datelist`를 정의합니다. 이 함수는 세 개의 매개변수 `start`, `end`, `end_month`를 받습니다.

2. `y_list = []`:
   - 빈 리스트 `y_list`를 생성합니다. 이 리스트는 나중에 날짜를 담을 용도로 사용됩니다.

3. `month_list = ['01', '02', '03', '04','05','06','07','08','09','10', '11', '12']`:
   - `month_list`에는 1월부터 12월까지의 두 자리 문자열 형태의 월이 들어 있는 리스트가 정의됩니다.

4. `for y in range(start, end):`:
   - `start`부터 `end-1`까지의 연도에 대해 반복합니다.

5. `if y != end-1:`:
   - 현재 연도 `y`가 마지막 연도(`end-1`)가 아닌 경우에만 해당합니다.

6. `for j in month_list:`:
   - `month_list`에 있는 각 월에 대해 반복합니다.

7. `y_value = str(y)+str(j)`:
   - 현재 연도 `y`와 월 `j`를 문자열로 결합하여 `y_value`에 저장합니다.

8. `y_list.append(y_value)`:
   - 생성된 날짜를 `y_list`에 추가합니다.

9. `else:`:
   - 현재 연도 `y`가 마지막 연도(`end-1`)인 경우 해당합니다.

10. `for j in month_list[:end_month]:`:
    - `end_month`까지의 월에 대해서만 반복합니다.

11. `y_value = str(y)+str(j)`:
    - 현재 연도 `y`와 월 `j`를 문자열로 결합하여 `y_value`에 저장합니다.

12. `y_list.append(y_value)`:
    - 생성된 날짜를 `y_list`에 추가합니다.

13. `return y_list`:
    - 최종적으로 생성된 날짜 리스트 `y_list`를 반환합니다.

이 함수는 `start`부터 `end-1`까지의 연도와 각 연도별로 1월부터 12월까지의 날짜를 문자열 형태로 생성하여 리스트에 담아 반환합니다. 예를 들어, `module_for_datelist(2020, 2022, 3)`를 호출하면 2020년부터 2021년까지의 1월부터 12월까지의 날짜와 2022년의 1월부터 3월까지의 날짜가 리스트로 반환됩니다.

```python
## 문자형으로 만들어진 month_list를 정수형 변수로 전환


## 필요한 변수는 동별 코드 (gu_code), 데이터 수집 월 (month_list)
def function_for_apt_price(gu_code, month_list):

    ServiceKey = KEYS['ODP']
    # month = 202106
    numOfRows = 5000

    data_list = []

    for mon in month_list:

        url = "http://openapi.molit.go.kr:8081/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTrade?LAWD_CD="+str(gu_code)+"&DEAL_YMD="+str(mon)+"&numOfRows="+str(numOfRows)+"&serviceKey="+str(ServiceKey)

        result = urlopen(url) #this is for response of XML
        house = BeautifulSoup(result, 'lxml-xml')
        te=house.find_all("item")

        #data_list = []

        for i in range (len(te)):
            price = te[i].거래금액.string.strip()
            built_yr = te[i].건축년도.string.strip()

            year = te[i].년.string.strip()
            month = te[i].월.string.strip()
            day = te[i].일.string.strip()

            dong_name = te[i].법정동.string.strip()
            apt_name = te[i].아파트.string.strip()
            size = te[i].전용면적.string.strip()
            gu_code = te[i].지역코드.string.strip()

            total = [price, built_yr, mon, year, month, day, dong_name, apt_name, size, gu_code]
            data_list.append(total)

        total_df = pd.DataFrame(data_list, columns= ["price", "built_yr", "Date", "year", "month", "day", "dong_name", "apt_name", "size", "gu_code"])

    total_df['price'] = total_df['price'].str.replace(',', '').astype(int)
    total_df['built_yr'] = total_df['built_yr'].astype(int)
    total_df['Date'] = pd.to_datetime(total_df['Date'], format="%Y%m")+ pd.offsets.MonthEnd(0)
    total_df['size'] = total_df['size'].astype(float)
    total_df['size_pyung'] = total_df['size']/3.33
    total_df['price_pyung'] = total_df['price']/total_df['size_pyung']
    total_df[['size_pyung',	'price_pyung']] = total_df[['size_pyung',	'price_pyung']].round(2)

    return total_df
```

### 3.3.2 함수화된 코드

- 날짜데이터를 만들어서 순자적으로 데이터를 수집하는 것이 관건
- 날짜를 형식에 맞추어 생성하는 함수를 따로 제작

```python
from datetime import datetime, timedelta

# 시작 날짜 설정 (2023년 1월 1일)
start_date = datetime(2023, 1, 1)

# 현재 날짜 구하기
end_date = datetime.today()

# 시작 날짜부터 오늘까지의 날짜 생성
current_date = start_date

date_list = []

while current_date <= end_date:
    formatted_date = current_date.strftime("%Y%m%d")
    date_list.append(formatted_date)
    current_date += timedelta(days=1)
```

```python
def get_movie_date(yr, mn, dt):

  # 시작 날짜 설정 (2023년 1월 1일)
    start_date = datetime(yr, mn, dt)

    # 현재 날짜 구하기
    end_date = datetime.today()

    # 시작 날짜부터 오늘까지의 날짜 생성
    current_date = start_date

    date_list = []

    while current_date <= end_date:
        formatted_date = current_date.strftime("%Y%m%d")
        date_list.append(formatted_date)
        current_date += timedelta(days=1)

    return date_list
```

```python
def get_movie_data(yr, mn, dt):

    dt_list = get_movie_date(yr, mn, dt)

    # 데이터 집계의 시차를 감안하여 일주일 여유를 둔다
    dt_list = dt_list[:-7]

    df_list = []

    movie_key = KEYS['MOVIE']

    for dt in dt_list:

        movie_url = f'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key={movie_key}&targetDt={dt}'

        response = urlopen(movie_url) #this is for response of json

        json_data = json.load(response)

        # 'dailyBoxOfficeList' 키에 해당하는 값을 가져옵니다.
        movie_data = json_data['boxOfficeResult']['dailyBoxOfficeList']

        # DataFrame을 생성합니다.
        df = pd.DataFrame(movie_data)

        df['date'] = dt

        df_list.append(df)

    df = pd.concat(df_list)

    return df
```

## 3.3.6 회귀식과 그 결과 평가

```python
import pandas as pd
import statsmodels.api as sm

# 데이터 프레임 생성
# (데이터를 여기에 넣어주세요)

# 독립변수 (X)와 종속변수 (y) 설정
X =  concated_df['salesShare']
y =  concated_df['salesAmt_y']

# 상수항 추가
X = sm.add_constant(X)

# 회귀 모델 피팅
model = sm.OLS(y, X).fit()

# 회귀 분석 결과 요약
print(model.summary())
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan

# 모델을 만들고 잔차를 계산
# 예시: 모델 = sm.OLS(y, X).fit()
residuals = model.resid

# 바톤 테스트
_, pval, _, _ = het_breuschpagan(residuals, model.model.exog)
print(f"P-value for Breusch-Pagan test: {pval}")
```

#### 회귀식 결과 해석

- 개봉일 3일의 시장점유율은 개봉후 20일의 흥행성과를 예측하는데 유의한 변수로 확인 (p-value <= 0.05) 됩니다.

해당 기타 회귀식 통계 결과의 해석은 다음과 같습니다:

1. Omnibus (오므니버스 검정):
   - 이 검정은 회귀 모델의 정규성 가정을 확인하는 테스트입니다.
   - 결과에서 Omnibus 값이 16.183이며, p-value가 0.000으로 유의수준 0.05보다 작습니다.
   - 따라서, 이 모델의 오차 항은 정규분포를 따르지 않는 것으로 나타났습니다.

2. Durbin-Watson (더빈-왓슨 통계량):
   - 이 통계량은 회귀 모델의 오차항 사이의 자기상관을 검정하는 통계량입니다.
   - Durbin-Watson 값이 2.360으로 나타났습니다.
   - 이 값이 2에 가까우면 자기상관이 존재하지 않는 것으로 해석됩니다.

3. Prob(Omnibus) (오므니버스 검정의 p-value):
   - 이 값이 0.000으로 유의수준 0.05보다 작습니다.
   - 따라서, 이 모델의 오차 항은 정규분포를 따르지 않는 것으로 나타났습니다.

4. Jarque-Bera (자크-베라 검정):
   - 이 검정도 회귀 모델의 정규성 가정을 확인하는 테스트입니다.
   - 결과에서 Jarque-Bera 값이 13.934이며, p-value가 0.000942로 유의수준 0.05보다 작습니다.
   - 따라서, 이 모델의 오차 항은 정규분포를 따르지 않는 것으로 나타났습니다.

5. Skew (왜도):
   - 왜도는 분포의 비대칭 정도를 나타냅니다.
   - 결과에서 Skew 값이 1.853으로 양수입니다. 이는 오른쪽으로 치우친 형태를 의미합니다.

6. Prob(JB) (자크-베라 검정의 p-value):
   - 이 값이 0.000942로 유의수준 0.05보다 작습니다.
   - 따라서, 이 모델의 오차 항은 정규분포를 따르지 않는 것으로 나타났습니다.

7. Kurtosis (첨도):
   - 첨도는 분포의 뾰족한 정도를 나타냅니다.
   - 결과에서 Kurtosis 값이 5.438으로 뾰족한 분포를 보이고 있습니다.

8. Cond. No. (Condition Number):
   - 다중공선성을 나타내는 지표입니다. 값이 크면 다중공선성의 가능성이 높습니다. 이 경우 65.2로 나타났습니다.

종합적으로, 해당 회귀 모델의 오차 항은 정규분포를 따르지 않으며, 다중공선성의 가능성이 있을 수 있습니다. 또한, 자기상관이 존재하지 않는 것으로 나타났습니다.

#### 잔차가 정규분포를 따르지 않을 경우의 신뢰성 문제

잔차가 정규분포를 따르지 않을 때 회귀식을 얼마나 신뢰할 수 있는지는 분석의 목적과 데이터의 특성에 따라 다를 수 있습니다.

1. **잔차의 정규성이 중요하지 않을 때**:
   - 종종, 회귀 분석에서 정규성 가정은 중요하지 않을 수 있습니다. 특히 큰 데이터셋이나 모델이 다른 가정을 충족할 때, 잔차가 정규분포를 따르지 않아도 모델이 유용한 예측을 할 수 있을 수 있습니다.

2. **중심 극한 정리를 활용할 때**:
   - 중심 극한 정리에 따르면, 샘플 크기가 크면 잔차의 분포가 정규분포에 가까워집니다. 따라서 대규모 데이터셋의 경우, 정규성 가정이 덜 중요할 수 있습니다.

3. **다른 가정들이 충족될 때**:
   - 다른 회귀 분석 가정들 (선형성, 등분산성 등)이 충족될 때, 잔차의 정규성이 덜 중요할 수 있습니다.

4. **예측의 정확성이 중요할 때**:
   - 모델의 예측 정확성이 주요 관심사인 경우, 잔차의 정규성은 상대적으로 중요할 수 있습니다.

5. **추가적인 검증을 통해 확인할 때**:
   - 잔차의 정규성을 확인하는 통계적 검증을 통해 모델의 신뢰성을 평가할 수 있습니다. QQ 플롯이나 잔차의 히스토그램을 통해 확인할 수 있습니다.

6. **다른 대안 모델이 더 적절할 때**:
   - 다른 회귀 모델이나 비모수적 방법이 더 적절한 경우가 있을 수 있습니다.

요약하자면, 잔차의 정규성이 중요하지 않을 수 있지만, 신뢰할 수 있는 회귀 모델을 구축하려면 다른 회귀 가정들을 충족해야 합니다. 또한, 잔차의 분포를 확인하고 다른 모델이나 접근 방법을 고려하는 것이 합리적일 수 있습니다.

**<참고 : 잔차가 정규분포를 따르지 않을 경우>**

잔차가 정규분포를 따르지 않는다면 몇 가지 대안적인 접근 방법이 있습니다. 이러한 상황에서는 다음과 같은 조치를 취할 수 있습니다:

1. **변수 변환 (Transformation)**:
   - 데이터의 변환을 통해 잔차를 정규분포에 가깝게 만들 수 있습니다. 예를 들어, 로그 변환, 제곱근 변환 등을 사용할 수 있습니다.

2. **가중회귀 (Weighted Regression)**:
   - 데이터 포인트에 가중치를 부여하여 모델을 학습할 수 있습니다. 이를 통해 이상치나 특이값의 영향을 줄일 수 있습니다.

3. **이상치 처리 (Outlier Handling)**:
   - 이상치를 식별하고 제거하거나, 다른 방법으로 처리할 수 있습니다.

4. **비모수적 방법 (Nonparametric Methods)**:
   - 비모수적 통계 기법을 사용하여 가정이 필요하지 않는 방법으로 모델링할 수 있습니다. 예를 들어, 부트스트랩 또는 로버스트 회귀 등을 사용할 수 있습니다.

5. **일반화 선형 모델 (Generalized Linear Models, GLM)**:
   - 정규분포 외의 다른 확률 분포를 가정하는 모델을 사용할 수 있습니다. 이를 통해 잔차의 분포를 더 잘 모델링할 수 있습니다.

6. **변수 추가 또는 제거**:
   - 모델에 새로운 변수를 추가하거나 불필요한 변수를 제거하여 모델의 적합성을 개선할 수 있습니다.

7. **비모수적 잔차 분석**:
   - QQ 플롯, 잔차의 히스토그램 등을 사용하여 잔차의 분포를 시각적으로 검토할 수 있습니다.

8. **모델 변경**:
   - 다른 모델을 시도하여 잔차의 분포를 더 잘 모델링할 수 있습니다. 예를 들어, 선형 회귀 대신 비선형 모델을 사용할 수 있습니다.

9. **로버스트 회귀 (Robust Regression)**:
   - 이상치에 덜 민감한 로버스트 회귀 모델을 사용할 수 있습니다.

이러한 방법들은 잔차의 정규성을 개선하고 모델의 적합성을 향상시킬 수 있습니다. 그러나 어떤 방법이 가장 적절한지는 데이터와 문제에 따라 다를 수 있으며, 실험적으로 확인하는 것이 좋습니다.
