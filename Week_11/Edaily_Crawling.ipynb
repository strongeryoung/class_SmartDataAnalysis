## URL정보 수집을 위한 완전 코드



```python
start_date = '20230101'
end_date = '20231115'
page_num = 3

url_head = 'https://www.edaily.co.kr'

url_lists = []


for i in range(10):

    print(i)

    url = f'https://www.edaily.co.kr/search/news/?source=ticon&keyword=%ec%a0%9c%ec%a3%bc%eb%8f%84+%ea%b4%80%ea%b4%91&include=&exclude=&jname=&start={start_date}&end={end_date}&sort=latest&date=pick&exact=false&page={i}'
    response = requests.get(url)
    contents = response.text
    soup = BeautifulSoup(contents, 'html.parser')

    link_info = soup.find_all('div', attrs={'class', 'newsbox_04'})

    print(link_info)

    link_num = len(link_info)
    print(link_num)

    i = 0  # 초기화

    while i < link_num:
        url_body = link_info[i].find('a')['href']
        url_links = url_head + url_body
        print(url_links)
        url_lists.append(url_links)
        i += 1
```

## 반복문을 사용해서 모든 기사정보 추출하기

```python
news_url = url_lists[:20]

aticle_data = []

for _url in news_url:

    response = requests.get(_url)
    contents = response.text
    soup = BeautifulSoup(contents, 'html.parser')

    main_text = soup.find('section', attrs={'class', 'aside_left'})

    title = main_text.find('div', attrs={'class', 'news_titles'}).get_text().replace('\n','' )

    if '등록' in title:
        tit = title.split('등록')[0]
    else:
        tit = title

    # 정규 표현식을 사용하여 "등록" 이후의 날짜 정보 추출
    match = re.search(r'등록 (\d{4}-\d{2}-\d{2})', title)

    if match:
        date_info = match.group(1)
        # date_txt.append(date_info)
    else:
        date_info = '미확인'

    main_text = main_text.find('div', attrs={'class', 'news_body'}).get_text().replace('\n','' )
    aticle_data.append((tit, date_info, main_text))

pd.DataFrame(aticle_data, columns=['제목', '날짜', '기사내용'])
```

```python
def clean_text(x):

    x = x.replace('\t', '')
    x = x.replace('\r', '')
    # [ ] 로 둘러싸인 부분 제거
    x = re.sub(r'\[.*?\]', '', x)

    return x
```

```python
news_df = pd.DataFrame(aticle_data, columns=['제목', '날짜', '기사내용'])
news_df['제목'] = news_df['제목'].apply(lambda x: clean_text(x))
news_df['기사내용'] = news_df['기사내용'].apply(lambda x: clean_text(x))
```
