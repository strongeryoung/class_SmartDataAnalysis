```python
url = 'https://www.ubase.co.kr/career/career_board.php'
response = requests.get(url)
contents = response.text
soup = BeautifulSoup(contents, 'html.parser')

info_list = []

ad_list = soup.find_all('div', attrs={'class', 'recom_list2'})
li_list = ad_list[0].find_all('li')
client_num = len(li_list)

for i in range(client_num):
    if i != client_num:
        client = li_list[i].find_all('div', attrs={'class', 'tit multiple2'})[0].get_text()
        location = li_list[i].find_all('div', attrs={'class', 'tag'})[0].get_text().replace('#', '') #location에 div class는 tag
        info_list.append((client, location))
    else:
        break
```

```python
post_info = pd.DataFrame(info_list, columns=['공고제목', '사업장위치'])
post_info['고객사'] = post_info['공고제목'].apply(lambda x: re.findall(r'\[(.*?)\]', x))
post_info['공고제목'] = post_info['공고제목'].apply(lambda x: re.sub(r'\[.*?\]', '', x))
# Print the extracted words
post_info['고객사'] = post_info['고객사'].apply(lambda x: ', '.join(x))
post_info['사업장위치'] = post_info['사업장위치'].apply(lambda x: x.replace('\n', ''))
post_info = post_info[['고객사', '공고제목', '사업장위치']]
today = date.today()
```
