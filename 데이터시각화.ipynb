## **5.1 Simple pyplot example**




````python
fig = plt.figure()
subplot = fig.add_subplot(1, 1, 1)
subplot.plot([1, 2, 3, 4])
plt.show()
````

해당 코드는 Matplotlib를 사용하여 간단한 그래프를 생성하는 예제입니다.

1. `fig = plt.figure()`:
   - 새로운 Figure 객체를 생성합니다. Figure 객체는 그림 전체를 나타냅니다.

2. `subplot = fig.add_subplot(1, 1, 1)`:
   - Figure에 서브플롯(축)을 추가합니다. 이 경우 1x1 그리드에서 첫 번째 위치에 서브플롯을 추가합니다. 즉, 하나의 서브플롯만 가지는 그림이 됩니다.

3. `subplot.plot([1, 2, 3, 4])`:
   - 서브플롯에 데이터를 그립니다. 여기서는 [1, 2, 3, 4] 라인 플롯을 생성합니다.

4. `plt.show()`:
   - 그림을 화면에 표시합니다.

결과적으로 이 코드는 하나의 서브플롯을 가지는 그림을 생성하고, 그 서브플롯에 [1, 2, 3, 4] 라인 플롯을 그립니다.

````python
fig = plt.figure(figsize=(10, 5))

subplot1 = fig.add_subplot(1, 3, 1)
subplot2 = fig.add_subplot(1, 3, 2)
subplot3 = fig.add_subplot(1, 3, 3)
subplot1.plot([1, 2, 3, 4])
subplot2.plot([4, 3, 2, 1])
subplot3.plot([4, 4, 5, 5])
plt.show()
````

이 코드는 Matplotlib를 사용하여 3개의 서브플롯을 가지는 그림을 생성하는 예제입니다.

1. `fig = plt.figure(figsize=(10, 5))`:
   - 새로운 Figure 객체를 생성합니다. `figsize=(10, 5)`는 그림의 크기를 가로 10인치, 세로 5인치로 지정합니다.

2. `subplot1 = fig.add_subplot(1, 3, 1)`:
   - Figure에 첫 번째 서브플롯을 추가합니다. 1행 3열의 그리드에서 첫 번째 위치에 서브플롯을 추가합니다. 즉, 1행 3열 중에서 첫 번째 열에 위치합니다.

3. `subplot2 = fig.add_subplot(1, 3, 2)`:
   - Figure에 두 번째 서브플롯을 추가합니다. 1행 3열의 그리드에서 두 번째 위치에 서브플롯을 추가합니다. 즉, 1행 3열 중에서 두 번째 열에 위치합니다.

4. `subplot3 = fig.add_subplot(1, 3, 3)`:
   - Figure에 세 번째 서브플롯을 추가합니다. 1행 3열의 그리드에서 세 번째 위치에 서브플롯을 추가합니다. 즉, 1행 3열 중에서 세 번째 열에 위치합니다.

5. `subplot1.plot([1, 2, 3, 4])`, `subplot2.plot([4, 3, 2, 1])`, `subplot3.plot([4, 4, 5, 5])`:
   - 각각의 서브플롯에 데이터를 그립니다. `subplot1`에는 [1, 2, 3, 4]를, `subplot2`에는 [4, 3, 2, 1]을, `subplot3`에는 [4, 4, 5, 5]를 라인 플롯으로 그립니다.

6. `plt.show()`:
   - 그림을 화면에 표시합니다.

결과적으로 이 코드는 가로로 3개의 서브플롯을 가지는 그림을 생성하고, 각각의 서브플롯에 데이터를 라인 플롯으로 그립니다.

````python
import FinanceDataReader as fdr

df = fdr.DataReader('000660','2021-01-01','2023-09-20')
df.rename(columns={"Close":"종가","Volume":"거래량"},inplace = True)
df.insert(5,'등락',df['종가'].diff())
df['5일_이동평균']=df['종가'].rolling(5).mean()
df['20일_이동평균']=df['종가'].rolling(20).mean()
df['60일_이동평균']=df['종가'].rolling(60).mean()
df['120일_이동평균']=df['종가'].rolling(120).mean()
df = df[['종가','등락','거래량','5일_이동평균','20일_이동평균','60일_이동평균','120일_이동평균']]

df
````

이 코드는 FinanceDataReader를 사용하여 한국 주식 종목 '000660'에 대한 주가 데이터를 가져와서 가공하는 작업을 수행합니다.

* FinanceDataReader는 7강 시간에 주로 다룰 예정임

1. `import FinanceDataReader as fdr`:
   - FinanceDataReader 라이브러리를 `fdr` 별칭으로 가져옵니다. 이 라이브러리는 금융 데이터를 쉽게 가져올 수 있는 도구입니다.

2. `df = fdr.DataReader('000660','2021-01-01','2023-09-20')`:
   - FinanceDataReader를 사용하여 '000660' 종목의 2021년 1월 1일부터 2023년 9월 20일까지의 주가 데이터를 가져와서 DataFrame 형식으로 `df`에 저장합니다.

3. `df.rename(columns={"Close":"종가","Volume":"거래량"},inplace = True)`:

4. `df.insert(5,'등락',df['종가'].diff())`:

   - `df['5일_이동평균'] = df['종가'].rolling(5).mean()` : 5일 이동평균을 계산하여 '5일_이동평균' 열에 추가합니다.
   - `df['20일_이동평균'] = df['종가'].rolling(20).mean()` : 20일 이동평균을 계산하여 '20일_이동평균' 열에 추가합니다.
   - `df['60일_이동평균'] = df['종가'].rolling(60).mean()` : 60일 이동평균을 계산하여 '60일_이동평균' 열에 추가합니다.

### **Simple Bar Chart**

````python
# 데이터
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 9]

# 막대 그래프 생성
plt.bar(categories, values)

# 그래프 제목과 레이블 추가
plt.title('예시 막대 그래프')
plt.xticks(fontsize= 12)
plt.yticks(fontsize= 12)
plt.xlabel('카테고리')
plt.ylabel('값')

# 그래프 표시
plt.show()
````

이 코드는 Python에서 Matplotlib 라이브러리를 사용하여 간단한 막대 그래프를 생성하고, 그래프의 제목과 레이블을 추가한 예시입니다.

1. **데이터 정의**:
   - `categories = ['A', 'B', 'C', 'D']`: `categories` 리스트에는 x축의 카테고리 레이블이 저장되어 있습니다.
   - `values = [3, 7, 1, 9]`: `values` 리스트에는 각 카테고리에 대응하는 y값이 저장되어 있습니다.

2. **막대 그래프 생성**:
   - `plt.bar(categories, values)`: `plt.bar()` 함수를 사용하여 막대 그래프를 생성합니다. `categories`는 x축의 레이블을 나타내고, `values`는 각 카테고리의 높이를 나타냅니다.

3. **그래프에 제목과 레이블 추가**:
   - `plt.title('예시 막대 그래프')`: 그래프에 "예시 막대 그래프"라는 제목을 추가합니다.
   - `plt.xlabel('카테고리')`: x축에 "카테고리"라는 레이블을 추가합니다.
   - `plt.ylabel('값')`: y축에 "값"이라는 레이블을 추가합니다.

4. **텍스트 크기 조절**:
   - `plt.xticks(fontsize= 12)`: x축 레이블의 폰트 크기를 12로 설정합니다.
   - `plt.yticks(fontsize= 12)`: y축 레이블의 폰트 크기를 12로 설정합니다.

5. **그래프 표시**:
   - `plt.show()`: 앞서 정의한 모든 설정을 바탕으로 그래프를 화면에 표시합니다.

이 코드를 실행하면 "A", "B", "C", "D"라는 네 가지 카테고리에 대한 막대 그래프가 생성되며, 그래프에는 제목과 각 축의 레이블이 추가됩니다. 또한, x축과 y축 레이블의 폰트 크기가 12로 설정됩니다.

````python
# 데이터
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 9]

# 막대 그래프 생성
plt.bar(categories, values)

# 각 막대에 값을 표시
for i, v in enumerate(values):
    plt.text(i, v + 0.02, str(v), ha='center', va='bottom', fontsize=10)

# 그래프 제목과 레이블 추가
plt.title('예시 막대 그래프')
plt.xlabel('카테고리')
plt.ylabel('값')

# 그래프 표시
plt.show()
````

막대 그래프 안에 값을 표시하기 위해서는 `plt.text()` 함수를 사용할 수 있습니다. 이 함수를 활용하여 각 막대 위에 값을 표시할 수 있습니다.

아래는 코드에 값을 추가한 버전입니다:

```python
import matplotlib.pyplot as plt

# 데이터
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 9]

# 막대 그래프 생성
plt.bar(categories, values)

# 각 막대에 값을 표시
for i, v in enumerate(values):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom', fontsize=10)

# 그래프 제목과 레이블 추가
plt.title('예시 막대 그래프')
plt.xlabel('카테고리')
plt.ylabel('값')

# 그래프 표시
plt.show()
```

여기서 `plt.text()` 함수를 사용하여 각 막대 위에 값을 추가했습니다. `enumerate(values)`를 통해 각 막대의 인덱스와 값을 순회하며, `plt.text(i, v + 0.1, str(v), ha='center', va='bottom', fontsize=10)`를 사용하여 값을 추가합니다.

- `i`: 막대의 인덱스
- `v`: 해당 막대의 값
- `v + 0.1`: 값이 표시될 y 좌표를 설정합니다. `0.1`은 조절값으로, 값이 막대 위에 어느 정도 떠 있을지 조절합니다.
- `str(v)`: 값을 문자열로 변환하여 표시합니다.
- `ha='center'`: 가로로 중앙 정렬
- `va='bottom'`: 세로로 아래 정렬
- `fontsize=10`: 텍스트의 폰트 크기를 10으로 설정합니다.

이제 각 막대 위에 해당 값이 표시됩니다.

3. **`plt.title('시가총액 상위 종목 파이 차트')`**:
   - 그래프에 제목을 추가합니다.

4. **`plt.show()`**:
   - 그래프를 화면에 표시합니다.

이 코드는 시가총액 상위 종목을 기반으로 파이 차트를 생성하고 시각적으로 나타내는데 사용됩니다.

````python
# 파이 차트 그리기
plt.figure(figsize=(10, 8))
plt.pie(market_cap['시가총액'], labels=market_cap['종목'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(market_cap))), wedgeprops={'width': 0.6})

# 제목 설정
plt.title('시가총액 상위 종목 파이 차트')

# 그래프 표시
plt.show()
````

가운데가 구멍이 뚫려있는 파이 차트를 그리려면 plt.pie() 함수에서 wedgeprops 매개변수를 사용하여 중심에 구멍을 뚫을 수 있습니다. 이를 위해 wedgeprops 매개변수에 {'width': 0.4}와 같은 속성을 설정하여 파이 차트의 중심 부분을 뚫을 수 있습니다.

이 코드는 주어진 데이터프레임을 다루기 위한 다양한 가공 작업을 수행합니다. 각각의 작업은 아래와 같이 설명됩니다:

1. **`raw_df['price'] = raw_df['price'].str.replace(',', '').astype(int)`**:
   - `price` 열의 값에서 쉼표(`,`)를 제거하고 문자열을 정수형으로 변환합니다. 쉼표는 숫자 표기법에서 천 단위 구분을 위해 사용됩니다.

2. **`raw_df['built_yr'] = raw_df['built_yr'].astype(int)`**:
   - `built_yr` 열의 값을 정수형으로 변환합니다. `built_yr`은 건물이 지어진 연도를 나타냅니다.

3. **`raw_df['Date'] = pd.to_datetime(raw_df['Date'], format="%Y%m")+ pd.offsets.MonthEnd(0)`**:
   - `Date` 열의 값을 년월 형식으로 변환하고, 그 뒤에 `pd.offsets.MonthEnd(0)`을 더하여 해당 월의 마지막 날짜를 얻습니다.

4. **`raw_df['size'] = raw_df['size'].astype(float)`**:
   - `size` 열의 값을 부동소수점형(float)으로 변환합니다. `size`는 건물의 크기를 나타냅니다.

5. **`raw_df['size_pyung'] = raw_df['size']/3.33`**:
   - `size`를 3.33으로 나누어서 `size_pyung` 열을 생성합니다. `size_pyung`은 제곱미터를 평(坪) 단위로 변환한 값을 나타냅니다.

6. **`raw_df['price_pyung'] = raw_df['price']/raw_df['size_pyung']`**:
   - `price`를 `size_pyung`로 나누어서 `price_pyung` 열을 생성합니다. `price_pyung`은 평당 가격을 나타냅니다.

7. **`raw_df[['size_pyung', 'price_pyung']] = raw_df[['size_pyung', 'price_pyung']].round(2)`**:
   - `size_pyung`과 `price_pyung` 열의 값을 소수점 둘째 자리까지 반올림합니다.

이렇게 가공된 데이터프레임은 원본 데이터에 대해 다양한 변환 작업을 수행한 결과입니다. 이를 통해 가독성이 높고 분석이 용이한 형태로 데이터를 가공할 수 있습니다.

### **시각화 (2) : 박스플롯 그리기**


1. **상자 (Box):**
   - 상자의 아래부분은 데이터의 25% 지점 (1사분위수, Q1) 을 나타냅니다.

   - 일반적으로 수염의 길이는 1.5*(IQR) 이내로 설정되며, IQR은 사분위수 범위입니다 (IQR = Q3 - Q1).

3. **이상치 (Outliers):**
   - 상자그림 밖에 위치한 데이터 포인트들을 이상치라고 합니다.

## **5.4 단일변수 회귀모형을 통한 가격 차별요인 분석**

````python
import statsmodels.api as sm

# 독립 변수와 종속 변수 설정
X = age_price['apt_age']
y = age_price['price_pyung']

# 상수항 추가
X = sm.add_constant(X)

# 회귀 모델 생성 및 피팅
model = sm.OLS(y, X).fit()

# 회귀 분석 결과 출력
print(model.summary())
````

#### 회귀분석 Code 해설

해당 코드는 주어진 데이터프레임 `age_price`에서 `apt_age`를 독립 변수로, `price_pyung`을 종속 변수로 설정하여 단순 선형 회귀 분석을 수행하는 코드입니다. 이를 각 단계별로 해설해보겠습니다:

1. **독립 변수와 종속 변수 설정**:
   - `X = age_price['apt_age']`: `apt_age` 열을 독립 변수 `X`로 설정합니다.
   - `y = age_price['price_pyung']`: `price_pyung` 열을 종속 변수 `y`로 설정합니다.

2. **상수항 추가**:
   - `X = sm.add_constant(X)`: 회귀 분석 모델에 상수항(intercept)을 추가합니다. 선형 회귀 모델에서는 주로 상수항을 포함하여 모델을 만듭니다. 이는 모델이 `y`축과의 절편을 가지게 만들어줍니다.

3. **회귀 모델 생성 및 피팅**:
   - `model = sm.OLS(y, X).fit()`: `sm.OLS`는 최소자승법(Ordinary Least Squares)을 이용한 선형 회귀 모델을 생성합니다. `y`는 종속 변수, `X`는 독립 변수를 의미합니다. `.fit()` 메서드를 사용하여 모델을 피팅합니다.

4. **회귀 분석 결과 출력**:
   - `print(model.summary())`: 회귀 분석 결과를 출력합니다. 이는 회귀 모델의 통계적 정보, 계수들의 유의성, 모델의 설명력 등을 요약해서 보여줍니다.

결과적으로, 위 코드는 주어진 데이터를 바탕으로 단순 선형 회귀 모델을 생성하고, 이 모델의 결과를 요약하여 출력합니다. 이를 통해 독립 변수와 종속 변수 간의 관계를 통계적으로 분석할 수 있습니다.

#### 회귀식 결과 분석

- apt_age : P_value < 0.05, 통계적으로 유의미한 변수

- 잔차 분석 결과

    - 자기상관성 검토

      - Durbin-Watson (DW) 통계량은 회귀 모델에서의 자기상관을 검정하는 지표입니다. 이 값은 일반적으로 0에서 4 사이의 값을 가질 수 있습니다.

        Durbin-Watson 통계량의 해석은 다음과 같습니다:

        - **DW < 2:** 양의 자기상관이 존재하는 것으로 해석됩니다. 즉, 잔차들 간에 양의 상관관계가 있다는 의미입니다. 이는 일반적으로 시계열 데이터에서 발생할 수 있습니다.

        - **DW = 2:** 자기상관이 없는 경우로 해석됩니다. 즉, 잔차들 간의 상관관계가 없다는 것입니다. 이는 회귀 모델의 가정 중 하나인 독립성을 나타냅니다.

        - **DW > 2:** 음의 자기상관이 존재하는 것으로 해석됩니다. 즉, 잔차들 간에 음의 상관관계가 있다는 의미입니다.

        Durbin-Watson 통계량이 1.653이면, 2에 근접하므로 자기상관이 거의 없다고 볼 수 있습니다. 이는 회귀 모델의 잔차들이 서로 독립적이라는 것을 나타냅니다.

    - 잔차의 정규성 검토

      - Jarque-Bera (JB) 검정은 잔차의 정규성을 검정하는 통계적 방법 중 하나입니다. JB 통계량은 정규분포를 따를 때 기대되는 값과 현재의 데이터에서 나타난 값 사이의 차이를 측정합니다.

        여기서 JB 검정 결과로 얻은 값이 0.321이라고 하였습니다. JB 통계량이 작으면 정규성 가정을 만족한다고 해석됩니다.

        일반적으로 JB 검정은 다음과 같은 가설을 검정합니다:

        - 귀무가설 (H0): 잔차는 정규분포를 따른다.
        - 대립가설 (H1): 잔차는 정규분포를 따르지 않는다.

        만약 p-value가 일정한 유의수준 (예를 들어, 0.05) 보다 크다면, 귀무가설을 기각할 충분한 증거가 없으므로, 잔차는 정규분포를 따른다고 할 수 있습니다. 반대로, p-value가 유의수준보다 작다면, 귀무가설을 기각하고 대립가설을 채택하게 됩니다.

        따라서 JB 통계량이 0.321이라는 매우 작은 값이면, 이는 잔차가 정규분포를 따른다는 증거가 강하다는 것을 나타냅니다.

    - 잔차 확률분포의 비대칭성과 첨도 검토

      `Skew`와 `Kurtosis`는 확률 분포의 비대칭성과 뾰족함을 나타내는 지표입니다.

        1. **Skew (비대칭도):** 확률 분포의 비대칭 정도를 나타내는 값입니다.
          - 0에 가까울수록 대칭에 가깝습니다.
          - 양수면 분포가 오른쪽으로 치우쳐져 있고, 음수면 왼쪽으로 치우쳐져 있습니다.

          여기서 Skew 값이 0.117이므로, 분포가 약간 오른쪽으로 치우쳐져 있습니다.

        2. **Kurtosis (첨도):** 확률 분포의 꼬리 부분의 뾰족함을 나타냅니다.
          - 3에 가까우면 정규분포와 비슷한 형태입니다.
          - 3보다 크면 뾰족한 꼬리를 가지며, 3보다 작으면 둥근 꼬리를 가집니다.

          여기서 Kurtosis 값이 2.999로, 정규분포와 거의 유사한 형태임을 나타냅니다.

        따라서 Skew와 Kurtosis 값이 각각 0.117과 2.999로 나왔다는 것은, 잔차의 분포가 정규분포에 근접하며, 큰 비대칭도나 뾰족함이 없음을 의미합니다.   




````python
# 예측값과 잔차 계산
y_pred = model.predict(X)
residuals = y - y_pred

# 잔차 분석 시각화
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# 잔차의 정규성 검정
sm.qqplot(residuals, line='r', ax=ax1)
ax1.set_title('QQ Plot')  # Q-Q Plot

# 잔차와 독립 변수 간의 관계 확인 (scatter plot)
ax2.scatter(X.loc[:, 'apt_age'], residuals)
ax2.axhline(y=0, color='r', linestyle='--')
ax2.set_xlabel('독립 변수')
ax2.set_ylabel('잔차')
ax2.set_title('Scatter plot of residuals vs. independent variable')

plt.tight_layout()
plt.show()
````

   - `sm.qqplot()`을 사용하여 잔차의 정규성을 시각적으로 확인합니다. QQ 플롯은 잔차가 정규분포를 따르는지를 확인합니다. `line='r'`은 QQ 플롯에서 빨간색 선을 그리도록 지정합니다.

8. `ax2.set_xlabel('독립 변수')` 및 `ax2.set_ylabel('잔차')`:
   - 두 번째 서브플롯의 x 축과 y 축에 레이블을 추가합니다.

   - 두 번째 서브플롯에 제목을 추가합니다.

    - 서브플롯 간의 간격을 조정하여 레이아웃을 보기 좋게 만듭니다.

    - 그림을 화면에 출력합니다.

````python
from scipy.stats import shapiro

# 모델 피팅 후 잔차 계산
# 예를 들어, model은 statsmodels에서 생성된 회귀 모델 객체입니다.
residuals = model.resid

# 샤피로-윌크 검정
stat, p_value = shapiro(residuals)

if p_value > 0.05:
    print("잔차는 정규분포를 따릅니다.")
else:
    print("잔차는 정규분포를 따르지 않습니다.")
````

````python
# 등분산성 검정
from statsmodels.stats.diagnostic import het_breuschpagan

_, p_value, _, _ = het_breuschpagan(residuals, X)
print(f"P-value: {p_value}")

# P-value를 이용하여 가설을 판단할 수 있습니다.
if p_value > 0.05:
    print("등분산성이 있다고 가설을 채택합니다.")
else:
    print("등분산성이 없다고 가설을 기각합니다.")
````

````python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(9, 5))

x = np.array(residuals)
mu = np.mean(x)
sigma = np.std(x)

x1 = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
sns.kdeplot(pd.DataFrame(residuals), fill=True, label='실제분포', ax=ax)
ax.plot(x1, stats.norm.pdf(x1, mu, sigma), color='red', label="정규분포")

ax.grid()
ax.legend()
ax.set_xlabel("")

plt.show()
````
